# Recurrent Neural Networks

In this notebook, we implement recurrent neural networks (RNNs) to forecast the number of Dengue cases one week ahead in three Brazilian cities. RNNs were developed to handle sequence data, such as time series. 



```python
import numpy as np
import pandas as pd

# plotting
import matplotlib.pyplot as plt
import seaborn as sns

# NN
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import kerastuner as kt
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor

# time series CV
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV

# evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## Long Short-Term Memory (LSTM)
One type of RNN architecture is Long Short-Term Memory (LSTM) developed by [Hochreiter and Schmidhuber (1997)](https://www.mitpressjournals.org/doi/abs/10.1162/neco.1997.9.8.1735).


A typical LSTM cell architecture is shown in below. 

<img src="lstm cell_unrolled.png">

Each LSTM cell contains four gates that determine how information flows during the forward-pass. Besides the input data vector at time $t$, $x_t$, the cell also receives as inputs the state of the previous cell $C_{t-1}$, and the hidden state vector $h_{t-1}$, which is the output of the previous cell. 
1. The forget gate $f_t$, decides which parts of the previous cell state $C_{t-1}$ will be deleted, i.e. "forgotten", and to which extent. The sigmoid activation layer of the forget gate generates values between 0 and 1 to determine which percentage of a piece of information will be forgotten. 
2. The input gate $i_t$ decides which parts of the previous cell state $C_{t-1}$ will be updated using new candidate values for the cell state $C'_t$. Again, the sigmoid activation layer determines whether the values are completely updated (value of 1), not updated at all (value of 0), or partially updated (value between 0 and 1). 
3. Finally, the output gate decides which parts of the new cell state $C_t$ will leave the cell as output that is also carried over to the next LSTM cell.

Specifically, we compute:

$$f_t = \sigma (W_f [h_{t-1}, x_t] + b_f)$$
$$i_t = \sigma (W_i [h_{t-1}, x_t] + b_i)$$
$$o_t = \sigma (W_o [C_{t}, h_{t-1}, x_t] + b_o)$$
$$C_t' = tanh (W_C [h_{t-1},x_t] +b_C)$$
$$C_t = f_t \cdot C_{t-1} + i_t \cdot C_t'$$
$$h_t = o_t \cdot tanh(C_t)$$

where $W_f, b_f, W_i, b_i, W_o, b_o, W_C, b_C$ are the weights and biases for the forget, input, output, and candidate gates, respectively.

```python
# setting the theme for our plots
sns.set_theme(style="darkgrid", palette='Set2', font='monospace')

# setting pandas output display limits
pd.set_option('display.max_rows', 500, 'display.max_columns', 500)

```

```python
# data setup

# read our data (train and test time series)
train_ts_norm = pd.read_csv('data/train_ts_norm.csv', parse_dates=['date']).sort_values('date', ascending=True)
test_ts_norm = pd.read_csv('data/test_ts_norm.csv', parse_dates=['date']).sort_values('date', ascending=True)


```

```python
# reshape for RNN
# input to our model needs to be of shape (batch, timesteps, features)
train_y = train_ts_norm.loc[:,'cases']
train_x = train_ts_norm.iloc[:,4:]

test_y = test_ts_norm.loc[:,'cases']
test_x = test_ts_norm.iloc[:,4:]

train_x_3d = np.reshape(train_x.values, (-1,9,1))
test_x_3d = np.reshape(test_x.values, (-1,9,1))
```

```python
# set up the RNN architecture

def model_builder(units, activation):
    model = keras.Sequential()
    model.add(layers.LSTM(units=units, activation=activation, input_shape=(train_x.shape[1],1)))
    model.add(layers.Dense(1))

    model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
                loss='mse')
                
    return model
```

```python
## HYPERPARAMETER TUNING

# we use an sklearn wrapper to be able to run the GridSearchCV with our time series split
lstm_model = KerasRegressor(build_fn=model_builder, epochs=100, batch_size=16, verbose=0)

# time series cross-validation (maintains time ordering)
n_splits = 2
tscv = TimeSeriesSplit(n_splits=n_splits, max_train_size=None)

# set up the parameter space
parameter_grid = {'units': [32, 64, 128],
                 'activation': ['relu', 'tanh']}

# grid search setup
clf = GridSearchCV(estimator = lstm_model, cv=tscv, param_grid=parameter_grid)

search_result = clf.fit(train_x_3d, train_y)
```

```python
search_result.best_params_
```

```python
search_result.best_score_
```

```python
# PREDICTIONS

# train set predictions
predictions_train = search_result.best_estimator_.predict(train_x_3d)

# test set predictions
predictions_test = search_result.best_estimator_.predict(test_x_3d)
```

```python
# combine the predictions
predictions_df = pd.DataFrame({'predictions':np.concatenate((predictions_train, predictions_test)),
                                 'actual': np.concatenate((train_y.values, test_y.values)),
                                 'city': np.concatenate((train_ts_norm.loc[:,'mun_name'], test_ts_norm.loc[:,'mun_name'])),
                                 'date': np.concatenate((train_ts_norm.loc[:,'date'], test_ts_norm.loc[:,'date']))
                                 })
```

```python
# INSPECT THE RESULTS

# plot
plt.figure(figsize=(20,5))

for i, city in enumerate(predictions_df.city.unique()):
    pred_city = predictions_df[predictions_df.city==city]
    plt.subplot(1,3,i+1)
    plt.plot(pred_city.date, pred_city.predictions, color='black', label='predicted')
    plt.plot(pred_city.date, pred_city.actual, color = 'green', label='actual')
    plt.title(city)
    plt.axvspan('2017-01-01', '2017-12-30', facecolor='yellow', alpha=0.2)
    plt.text(s='Test Set', x='2017-03-01', y=max(pred_city.actual.append(pred_city.predictions)), color = 'gray', fontweight='bold')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# compute error metrics

# mae
mae =  mean_absolute_error(test_y, predictions_test)
# rmse
rmse = mean_squared_error(y_true=test_y, y_pred=predictions_test, squared=False)

error_metrics = pd.DataFrame({'mae': mae, 
                             'rmse': rmse}, index=['errors_lstm'])
round(error_metrics,3)
```
